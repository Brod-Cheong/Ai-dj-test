```python

# Authentication + drive connection
from google.colab import userdata
from google.colab import drive
# drive.mount('/content/drive')
from openai import OpenAI

# Ensure that your API key is set as a 'secret' - under the left side bar "key" logo
openai_key_name = 'Open_AI_key'

def read_and_obfuscate_key(key_name):
  print(f"Reading {key_name} from secrets")
  key_value = userdata.get(key_name)
  if key_value:
  # If the API key is long enough, mask all but the first 4 and last 4 characters.
    if len(key_value) > 8:
        masked_key = key_value[:4] + '*' * (len(key_value) - 8) + key_value[-4:]
    else:
        # For keys 8 characters or shorter, simply mask the entire key.
        masked_key = '*' * len(key_value)
    print(f"{key_name}: {masked_key}")
  else:
    print(f"{key_name}: not found!")

  return key_value

openai_key = read_and_obfuscate_key(openai_key_name)
```

## Infinite Chatbot (with Sidecar)
_using another summarizer to deal with excessively long contexts_



We introduce a "Double" System now where there 2 chatbots.
- Main "professor" bot
  - **Input:** Summary + new student message
  - **Output:** New Professor message
- Sidecar "TA" bot
  - **Input:** Current Summary + new student message + new Professor message
  - **Output:** New Summary


```python
# Create a while loop - asking user for input at each time. Append the user message into messages along with the openai chat's response to create a conversation

chat_bot_response = "Hi there - how can I help you?"

chat_summarizer_prompt = '''
You are an teaching assistant assisting a professor with an open office hours session.
Please help keep track of the conversation between the student and the professor and summarize the key points.
These key points would be used by the professor to answer subsequent questions from the student.
So it is imperative that it captures all the important information but does not include duplicate/irrelevant information
'''



# Initializing original messages / "memory"
chat_summary = ""
student_message = ""
prof_message = ""

chat_summarizer_messages = [{"role": "system", "content": chat_summarizer_prompt}]

# Professor Chatbot:
# INPUT: Takes in summary + new student input
# OUTPUT: 1. INPUT to summarizer alongside student response + displayed to user

# TA Chatbot:
# Input: Takes in previous summary, new student input, new prof response
# OUTPUT: New summary

prof_message = "Hey there! How can I help you today?"

count = 0
MAX_COUNT = 5

while count < MAX_COUNT:
  count += 1
  print("\n\n\n")

  # Get user input + persist into chat history

  print(f'Student Message:\n {student_message}\n\n\n')
  print(f'Professor Message:\n {prof_message}\n\n\n')
  print(f'Current Chat Summary:\n {chat_summary}\n\n\n')

  # Student Response
  student_message = input("Next Student Message: ")

  # Prof Response
  prof_response = openai_client.chat.completions.create(
      model="gpt-4o-mini",
      messages=[
          {
              "role": "system", "content": chat_system_prompt + " Here is a summary of the existing conversation so far: " + chat_summary
          },
          {
              "role": "user", "content": student_message
          }
      ]
  )

  prof_message = prof_response.choices[0].message.content

  chat_summarizer_input = f'Existing summary: {chat_summary} - New student input: {student_message} - New professor input: {prof_message}'

  # Summarizer
  summarizer_response = openai_client.chat.completions.create(
      model="gpt-4o-mini",
      messages=[{"role": "system", "content": chat_summarizer_prompt}, {"role": "user", "content": chat_summarizer_input}]
  )

  chat_summary = summarizer_response.choices[0].message.content
```

## Base Evaluation Framework Set-Up

### Main intuition###
Capture the entire conversation from start to end, to check for "dissonance" - parts of the conversation that does not flow properly.

Note that we are okay if the "student" is the source of the dissonance (we cannot control humans), but if the professor (chatbot) response is dissonant to the chat history, then it should be penalized.

###Implementation###
Track the chat history between teacher and user on top of the previous code scaffold - pass the entire chat history to another LLM for evaluation afterwards

```python
# Create a while loop - asking user for input at each time. Append the user message into messages along with the openai chat's response to create a conversation

chat_history = []

chat_bot_response = "Hi there - how can I help you?"

chat_summarizer_prompt = '''
You are an teaching assistant assisting a professor with an open office hour.
Please help keep track of the conversation between the student and the professor and summarize the key points.
These key points would be used by the professor to answer subsequent questions from the student. So it is imperative that it captures all the important information but does not include duplicate/irrelevant information
'''

chat_summary = ""
student_message = ""
prof_message = ""


chat_summarizer_messages = [{"role": "system", "content": chat_summarizer_prompt}]

prof_message = "Hey there! How can I help you today?"

while True:

  # Displaying Professor's response
  print(prof_message)

  # Student Response
  student_message = input("User Input: ")

  # POPULATING NEW CHAT_HISTORY LIST
  chat_history.append({"role": "student", "message": student_message})

  # Prof Response
  prof_response = openai_client.chat.completions.create(
      model="gpt-4o-mini",
      messages=[
          {
              "role": "system", "content": chat_system_prompt + " Here is a summary of the existing conversation so far: " + chat_summary
          },
          {
              "role": "user", "content": student_message
          }
      ]
  )

  prof_message = prof_response.choices[0].message.content

  # POPULATING NEW CHAT_HISTORY LIST
  chat_history.append({"role": "professor", "message": prof_message})

  chat_summarizer_input = f'Existing summary: {chat_summary} - New student input: {student_message} - New professor input: {prof_message}'

  # Summarizer
  summarizer_response = openai_client.chat.completions.create(
      model="gpt-4o-mini",
      messages=[{"role": "system", "content": chat_summarizer_prompt}, {"role": "user", "content": chat_summarizer_input}]
  )

  chat_summary = summarizer_response.choices[0].message.content
```

## Using Structured Data for Software Integration / For Data Analysis
To integrate LLM flows/power seamlessly into software, we need data structures/objects that are consistent

In this case, let's create an "Audit" object that can capture that information

```python
# Pydantic Classes/Enum to structure LLM output

class Score(str, Enum):
  clear = "CLEAR"
  slightly_dissonant = "SLIGHTLY_DISSONANT"
  very_dissonant = "VERY_DISSONANT"
  unclear = "UNCLEAR"

class Audit(BaseModel):
    score: Score
    justification: str
```

```python
class SarcasmEnum(str, Enum):
  sarcastic = "SARCASTIC"
  not_sarcastic = "NOT_SARCASTIC"

# BaseModel
class SarcasmResult(BaseModel):
  sarcasm: SarcasmEnum
  justification: str

def check_sarcasm(review_content):
  response = openai_client.beta.chat.completions.parse(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "system",
            "content": "You are a sarcasm evaluator. Please evaluate the below review to determine if it is sarcastic or not."
        },
        {
            "role": "user",
            "content": [
                {"type": "text", "text": review_content}
            ]
        }
    ],
    response_format=SarcasmResult
  )

  sarcasm_result = response.choices[0].message.parsed
  return sarcasm_result.sarcasm.value

print(check_sarcasm("This charger is super useful and amazing, it spoilt only within the first week!"))
```

```python
df_subset['sarcasm_result'] = df_subset['review_content'].apply(check_sarcasm)
```
